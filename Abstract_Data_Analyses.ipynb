```python
# safe divide function to stop zero counts from causing problems
def safe_divide(a, b):
    if b != 0: #
        return a/b
    else:
        return 0
```

```python
# Create a list of causal connectives
CC = ["although", "arise", "arises", "arising", "arose", "because", "cause", "caused", "causing", "consequence", "consequences", "consequent", "consequently", "since", "due", "enable", "enabled", "enables", "given", "therefore", "thus", "hence", "accordingly", "for this reason", "on account of", "in light of", "in view of", "in consequence of", "resulting in", "leads to", "as a result", "thereby", "in turn", "even then", "follow that", "follow the", "follow this", "following that", "follows the", "hence", "made", "make", "makes", "making", "nevertheless", "nonetheless", "only if", "provided that", "therefore", "though", "thus", "unless", "whenever"
]

# Create a list of implicit causal verbs
IC = [
    "abandon", "abash", "abhor", "acclaim", "accompany", "accuse", "admire", "admonish", "adore", "advise",
    "affect", "affront", "aggravate", "agitate", "alarm", "alienate", "amaze", "amuse", "anger", "annoy", "answer",
    "antagonize", "apologize to", "appall", "appease", "applaud", "appreciate", "approach", "astonish", "astound",
    "attract", "avoid", "baffle", "banish", "battle", "beguile", "believe", "betray", "bewilder", "blame", "bless",
    "bore", "bother", "bug", "call", "calm", "calm down", "captivate", "caress", "carry", "castigate", "catch",
    "caution", "celebrate", "censure", "charm", "chase", "chasten", "chastize", "cheat", "cheer", "cherish", "chide",
    "chill", "comfort", "commend", "compensate", "complement", "compliment", "concern", "condemn", "confess to",
    "confide in", "confound", "confuse", "congratulate", "console", "consult", "correct", "corrupt", "counsel",
    "court", "criticize", "cuddle", "date", "daunt", "debate with", "deceive", "decry", "defame", "defy", "delight",
    "denigrate", "denounce", "deplore", "deprecate", "deride", "desert", "despise", "detest", "disappoint",
    "discourage", "disdain", "disgruntle", "dislike", "disobey", "disparage", "distract", "distress", "distrust",
    "divorce", "dominate", "dread", "dream about", "dumbfound", "echo", "embrace", "employ", "encourage", "enlighten",
    "enliven", "enrage", "enthral", "entice", "entrance", "envy", "escort", "esteem", "exalt", "exasperate", "excite",
    "excuse", "exhaust", "exhilarate", "fancy", "fascinate", "favour", "faze", "fear", "feed", "film", "flabbergast",
    "flatter", "floor", "follow", "fool", "forgive", "forget", "fight", "free", "frighten", "frustrate", "gall",
    "gladden", "grab", "graze", "greet", "grieve", "guide", "hail", "harass", "harm", "hate", "haunt", "help", "hire",
    "hit", "honour", "hug", "hurt", "idolize", "incense", "infuriate", "inspire", "instruct", "insult", "interrupt",
    "intimidate", "intrigue", "invigorate", "irritate", "jollify", "jolt", "kill", "kiss", "laud", "laugh at", "lead",
    "leave", "lie to", "like", "loathe", "love", "madden", "marry", "mesmerize", "meet", "miss", "mock", "mollify",
    "mourn", "move", "nettle", "notice", "nuzzle", "order around", "pacify", "pain", "pardon", "pass", "peeve",
    "penalize", "persecute", "pet", "pick up", "pity", "placate", "plague", "play", "play with", "please", "praise",
    "prize", "prosecute", "protect", "provoke", "punish", "pursue", "question", "reassure", "rebuke", "recompense",
    "relax", "relish", "remunerate", "repay", "repel", "reprimand", "reproach", "reprove", "repulse", "resent",
    "respect", "revere", "revile", "revitalize", "revolt", "reward", "ridicule", "rush to", "salute", "scare", "scold",
    "scorn", "shadow", "shame", "shock", "shake", "sicken", "slander", "snub", "spank", "spook", "stagger", "stare at",
    "startle", "stimulate", "strike", "sue", "support", "surprise", "tail", "tantalize", "telephone", "thank", "toast",
    "tolerate", "take away", "torment", "track", "trail", "treasure", "trouble", "trust", "unnerve", "unsettle",
    "uplift", "upset", "value", "venerate", "vex", "victimize", "vilify", "visit", "want", "warn", "weary", "welcome",
    "worry", "worry about", "worship", "wound", "wow", "yearn for", "yell at"
]


# Create empty lists
nw_final = []
CC_final = []
advcl_final = []
dobj_final = []
sconj_final = []
IC_final = []

for index, row in abstracts_clean.iterrows(): #going through the dataframe by row
    nw = 0 #starting count for number of words
    cc = 0 #starting count for number of causal conjunctions
    advcl = 0 #starting count for adverbial clauses
    dobj = 0 #starting count for direct object
    sconj = 0
    ic = 0
    text = str(row['text '])  # Converting column 'text' to string so spacy can process it
    text = text.lower() #lower case because I've learned my lesson
    tokenized_doc = nlp(text)  #Spacy processes the text
    has_causal_connective = False  # A flag to track if a causal connective has been encountered
    has_implicit_causality = False
    for token in tokenized_doc:  # go through each token
        #print(token.lemma_, token.tag_, token.pos_)
        if not token.is_punct and not token.tag_ == "SPACE":
            nw += 1  # + word count for each non-punctuation token
        if token.tag_ == "IN" and token.text in CC: #if token tag is IN and in CC, count it
            #print(token.text, token.pos_, token.tag_, token.dep_, token.head)
            cc += 1  # + causal connective count
            has_causal_connective = True
        if  token.pos_ == "VERB" and token.lemma_ in IC: # if token pos is verb and the lemma is in IC, count it
            #print(f'this is the causal verb: {token.lemma_}')
            ic += 1  # + causal connective count
            has_implicit_causality = True
        if token.dep_ == "advcl" and has_causal_connective:
            #print(token.text, token.pos_, token.tag_, token.dep_, token.head)
            advcl += 1  # + advcl count
            has_causal_connective = False  # Reset the flag
        if token.dep_ == "dobj":
            #print(token.text, token.pos_, token.tag_, token.dep_, token.head)
            dobj += 1 #+ dobj count
        if token.pos_ == "SCONJ":
            #print(token.pos)
            sconj += 1 #add 1 to sconj count for each token

    nw_final.append(nw) #append number of words by text to list above
    CC_final.append(safe_divide(cc, nw) * 1000) #append, apply safe divide, and multiply by 1000 for interpretablity
    advcl_final.append(safe_divide(advcl, nw) * 1000)
    dobj_final.append(safe_divide(dobj, nw) * 1000)
    sconj_final.append(safe_divide(sconj, nw) * 1000)
    IC_final.append(safe_divide(ic, nw) * 1000)



print(nw_final)
print(CC_final)
print(advcl_final)
print(dobj_final)
print(sconj_final)
print(IC_final)
```

```python
#selecting columns/variables needed for logistic regression
selected_columns = ['Binary_design', 'CC', 'SCONJ', 'ADVCL', 'DOBJ','IC']  # Replace with your desired column names

# Using loc to select specific columns by name
abstracts_log = abstracts_clean.loc[:, selected_columns]

abstracts_log
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import statsmodels.api as sm

# Split the dataset into features and target variable
X = abstracts_log.drop("Binary_design", axis=1)
y = abstracts_log["Binary_design"]

# Standardize (scale) the feature data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create and fit a logistic regression model using statsmodels
X_scaled = sm.add_constant(X_scaled)  # Add a constant term for the intercept
model = sm.Logit(y, X_scaled)
result = model.fit()

# Print the summary including p-values
print(result.summary())

# Use the trained model to make predictions on the entire dataset
y_pred_prob = result.predict(X_scaled)  # Predict probabilities
y_pred = np.round(y_pred_prob)  # Convert probabilities to binary predictions (0 or 1)

# Evaluate the model
accuracy = accuracy_score(y, y_pred)
confusion_mat = confusion_matrix(y, y_pred)
classification_rep = classification_report(y, y_pred)

# Print evaluation metrics
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", confusion_mat)
print("Classification Report:\n", classification_rep)
```
